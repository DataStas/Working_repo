```python
from math import gcd
from functools import reduce
class Solution:
    def hasGroupsSizeX(self, deck) -> bool:
        groups = {}
        unique_cards = set(deck)
        for number in deck:
            if number in groups:
                groups[number] += 1
            else:
                groups[number] = 1
        cards_counter = []
        for number in groups.keys():
            cards_counter.append(groups[number])
        return True if reduce(gcd, cards_counter) != 1 else False
```
