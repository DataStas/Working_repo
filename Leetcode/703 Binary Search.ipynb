```python
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        l = 0 
        r = len(nums) - 1
        while r > l:
            m = l + (r-l+1)//2
            if (target < nums[m]):
                r = m - 1
            else:
                l = m
        return l if nums[l] == target else -1
```
