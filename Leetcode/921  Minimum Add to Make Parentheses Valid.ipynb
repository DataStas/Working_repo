```python
class Solution:
    def minAddToMakeValid(self, s) -> int:
        pattern = {'(': ')'}
        ind = 0
        for bracket in s[1:]:
            print(bracket, ind)
            if bracket != ')':
                ind += 1 
            else:
                ind -= 1
        return ind
```
