```python
import numpy as np
```

```python
grid = [[0,1]]
ans = 0
```

```python
my_grid = [0 for _ in range(len(grid)+2)]
for num in range(1, len(my_grid)-1): 
    if grid[num-1] == 1:
        my_grid[num] = grid[num-1]
for num, point in enumerate(my_grid):
    if point == 1:
        if my_grid[num-1] == 0:
            ans += 2
        if my_grid[num+1] == 0:
            ans += 2
```

```python
my_grid = [[0 for _ in range(len(grid)+2)] for _ in range(len(grid)+2)]
for num_row in range(1, len(my_grid)-1):
    for num_col in range(1, len(my_grid)-1):
        try: 
            if grid[num_row-1][num_col-1] != 0:
                my_grid[num_row][num_col] = grid[num_row-1][num_col-1]
        except TypeError:
            my_grid[num_row][num_col] = grid[num_row-1]
for num_row, row in enumerate(my_grid):
    for num_col, point in enumerate(row):
        if point == 1:
            if my_grid[num_row-1][num_col] == 0:
                ans += 1
            if my_grid[num_row][num_col-1] == 0:
                ans += 1
            if my_grid[num_row][num_col+1] == 0:
                ans += 1
            if my_grid[num_row+1][num_col] == 0:
                ans += 1
```

```python
class Solution:
    def islandPerimeter(self, grid: List[List[int]]) -> int:
        visit = set()

        def dfs(i, j):
            if i >= len(grid) or j >= len(grid[0]) or i < 0 or j < 0 or grid[i][j] == 0:
               return 1
            if (i, j) in visit:
                return 0
            
            visit.add((i, j))
            perim = dfs(i, j + 1)
            perim += dfs(i + 1, j)
            perim += dfs(i, j - 1)
            perim += dfs(i - 1, j)
            return perim
        
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == 1:
                    return dfs(i, j)
```
