```python
class Solution:
    def kthDistinct(self, arr: List[str], k: int) -> str:
        let_map = {}
        for letter in arr: 
            if letter in let_map:
                let_map[letter] += 1
            else:
                let_map[letter] = 1
        valid_chars = []
        for key, value in let_map.items():
            if value == 1:
                valid_chars.append(key)
        if len(valid_chars) < k:
            return ""
        else:
            return valid_chars[k-1]
```

```python
class Solution:
    def kthDistinct(self, arr: List[str], k: int) -> str:
        l = []
        x = Counter(arr)
        for i, j in x.items():
            if j==1:
                l.append(i)
        if len(l)<k:
            return ""
        return l[k-1]
```
