```python
class MyHashSet:

    def __init__(self):
        self.data = []
        

    def add(self, key: int) -> None:
        if key in self.data:
            pass
        else:
            self.data.append(key)
        return self.data
        

    def remove(self, key: int) -> None:
        if key in self.data:
            self.data.remove(key)
        

    def contains(self, key: int) -> bool:
        if key in self.data:
            return True
        else:
            return False
```
