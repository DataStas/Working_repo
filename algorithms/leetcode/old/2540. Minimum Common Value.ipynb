```python
nums1 = [12,16,24,24,25,27,31,37,38,41,43,50,57,70,71,71,74,76,77,78]
nums2 = [5,5,9,11,12,17,20,34,36,51,61,68,70,79,85,87,88,90,91,97]
```

```python
nums1 = tuple(set(nums1))
nums2 = tuple(set(nums2))
```

```python
if nums1[0] > nums2[0]:
    min_num = nums2[0]
    start = round(len(nums1)/2) - 1 
    num = -1
    while num != min_num:
        if nums1[start] > min_num:
            start -= round(start/2) - 1 
        else:
            start += round(start/2) - 1 
else:
    min_num = nums1[0]
    start = round(len(nums2)/2) - 1 
    num = -1
    while num != min_num:
        if nums2[start] > min_num:
            start -= round(start/2) - 1 
        else:
            start += round(start/2) - 1 
```

```python
n1, n2 = len(nums1), len(nums2)
i, j = 0, 0
while i < n1 and j < n2:
    if nums1[i] == nums2[j]:
        break
        # return nums1[i]
    if nums1[i] < nums2[j]:
        i += 1
    else:
        j += 1
```
