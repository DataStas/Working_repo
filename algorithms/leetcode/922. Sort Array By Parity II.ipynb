```python
class Solution:
    def sortArrayByParityII(self, nums):
        ans = []
        storage = {'odd':[], 'even':[]}
        for num in nums:
            if num % 2 == 0:
                storage['odd'] += [num]
            else:
                storage['even'] += [num]
        print(storage['odd'])
        for ind in range(len(nums)):
            if ind % 2 == 0:
                ans.append(storage['odd'][0])
                storage['odd'].remove(storage['odd'][0])
            else:
                ans.append(storage['even'][0])
                storage['even'].remove(storage['even'][0])
        return ans
```
