```python
nums = [4,4,2,4,3]
triplets = set()

def find_trip(i, j, k):
    if nums[i] != nums[j] and nums[j] != nums[k] and nums[i] != nums[k] and (i, j, k) not in triplets:
        triplets.add((i, j, k))


for i in range(len(nums)-2):
    for j in range(i+1, len(nums)-1):
        for k in range(j+1, len(nums)):
            print(i, j, k)
            find_trip(i, j, k)
```

```python
from collections import Counter
c = Counter(nums)
res = 0

left = 0
right = len(nums)

for _, freq in c.items():
    right -= freq
    res += left * freq * right
    left += freq
```

```python
c = {}
for num in nums:
    if num not in c:
        c[num] = 1
    else:
        c[num] += 1
res = 0

left = 0
right = len(nums)

for _, freq in c.items():
    right -= freq
    res += left * freq * right
    left += freq
```
