```python
def diStringMatch(self, s: str) -> List[int]:
    l = 0
    r = len(s)
    res = []
    for i in range(0, len(s), 2):
      sub = s[i:i+2]
      match sub:
        case "ID":
          res.extend([l, r])
          l += 1
          r -= 1
        case "DI":
          res.extend([r, l])
          r -= 1
          l += 1
        case "II" | "I":
          res.extend([l, l+1])
          l += 2
        case "DD" | "D":
          res.extend([r, r-1])
          r -= 2
    if len(res) < len(s)+1:
      res.append(l if s[-1] == "I" else r)
    return res
```
